```python
import sklearn as sk
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
retorno = load_iris()
```

```python
retorno
```

```python
#Classificando os dados entre setosa, versicolor e virginica
def defineNomes(target):
  if(target == 0):
    return 'setosa'
  elif(target == 1):
    return 'versicolor'
  elif(target == 2):
    return 'viriginica'
```

```python
#colocando os dados classificados num dataframe
iris = pd.DataFrame(retorno.data)
iris.columns = ['comprimento da sépala', 'largura da sépala', 'comprimento da pétala', 'largura da pétala']
```

```python
#Adicionando as colunas target
iris['target'] = retorno.target
iris['target_names'] = iris['target'].apply(defineNomes)
```

```python
display(iris)
```

```python
iris.describe()
```

```python
sns.boxplot(data=iris)
```

```python
#verificando os outliers
iris[(iris['largura da sépala'] < 2.2) | (iris['largura da sépala'] > 4)]
```

```python
#verificando valores nulos
iris.info()
```

```python
#correlação entre os dados
#correlação = 1, tem uma correlação forte, onde uma cresce e a outra cresce na mesma proporção
#correlação = -1, tem uma correlação forte, onde uma cresce e a outra decresce na mesma proporção
#correlação = 0, não há correlação
iris.corr()
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue = 'target_names')
```

```python
sns.scatterplot(x='largura da pétala', y='comprimento da pétala', data=iris, hue='target_names', palette=['green','orange','blue'])
```

Separando dados em treino e teste

```python
from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split(retorno.data, retorno.target, random_state=1)
```

```python
#Criando data frame com modelo tde treino
treino = pd.DataFrame(x_treino)
treino.columns = ['comprimento da sépala','largura da sépala','comprimento da pétala','largura da pétala']
treino['target'] = y_treino
```

```python
display(treino)
```

```python
sns.scatterplot(x='largura da pétala', y = 'comprimento da pétala',data=treino, hue='target', palette=['green', 'red','blue'] )
```

```python
#verificando os pontos classificados errados
treino[(treino['target'] == 2) & ((treino['largura da pétala'] == 1.7) | (treino['largura da pétala'] == 1.5))]
```

```python
#traçando retas para separar os pontos
#mesmo errando 3 pontos a acertividade do modelo é de 90%
sns.scatterplot(x='largura da pétala', y='comprimento da pétala', data=treino, hue='target',palette=['green','red','blue'])
x_verde = [0,1.4]
y_verde = [4.8,0]
sns.lineplot(x=x_verde, y=y_verde, color='g')
x_azul = [0,4.9]
y_azul = [7.7,0]
sns.lineplot(x=x_azul, y=y_azul, color='b')
```

Tratando modelos de teste

```python
teste = pd. DataFrame(x_teste)
teste.columns = ['comprimento da sépala','largura da sépala','comprimento da pétala', 'largura da pétala']
teste['target'] = y_teste
```

```python
display(teste)
```

```python
#plot modelo de teste
sns.scatterplot(x='largura da pétala', y='comprimento da pétala', data=teste, hue='target',palette=['green','red','black'])
x_verde = [0,1.4]
y_verde = [4.8,0]
sns.lineplot(x=x_verde, y=y_verde,color='g')
x_preto = [0,4.9]
y_preto = [7.7,0]
sns.lineplot(x=x_preto, y=y_preto, color='black')
```
