```python
colunas_interesse = ['feridos_leves', 'ilesos', 'ignorados', 'feridos_graves', 'veiculos']
dados = data[colunas_interesse]

# Calculo o IQR para cada coluna
Q1 = dados.quantile(0.25)
Q3 = dados.quantile(0.75)
IQR = Q3 - Q1

# Calculo os limites superior e inferior para identificar outliers
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Identificando os outliers em cada coluna
outliers = ((dados < limite_inferior) | (dados > limite_superior)).any(axis=1)

dados_com_outliers = data[outliers]

print("Dados com outliers:")
print(dados_com_outliers)

#limites inferior e superior
print("Limite Inferior:")
print(limite_inferior)
print("Limite Superior:")
print(limite_superior)
```

```python
import numpy as np
# Aplicando uma transformação logarítmica para as colunas com outliers
columns_to_transform = ['feridos_leves', 'ilesos', 'ignorados', 'feridos_graves', 'veiculos']

for column in columns_to_transform:
    data[column] = np.log1p(data[column])
```

# K-Means


```python
data.info()
```

```python
# Substituindo vírgulas por pontos e convertendo para float para a coluna 'longitude'
data['latitude'] = data['latitude'].str.replace(',', '.').astype(float)
data['longitude'] = data['longitude'].str.replace(',', '.').astype(float)
```

```python
# Selecionando as colunas relevantes
data = data[['ilesos', 'feridos_leves', 'feridos_graves', 'mortos', 'latitude', 'longitude', 'uf']]

data['latitude'] = data['latitude'].astype(float)
data['longitude'] = data['longitude'].astype(float)
```

```python
from sklearn.preprocessing import StandardScaler
numeric_cols = ['ilesos', 'feridos_leves', 'feridos_graves', 'mortos', 'latitude', 'longitude']
categorical_cols = ['uf']

# Aplicando StandardScaler nas variáveis numéricas
scaler = StandardScaler()
data[numeric_cols] = scaler.fit_transform(data[numeric_cols])

# Reconcatenando as colunas categóricas
data = pd.concat([data, data[categorical_cols]], axis=1)
```

```python
from sklearn.cluster import KMeans
data = pd.get_dummies(data, columns=['uf'], prefix='uf')
```

```python
import plotly.express as px
inertia = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(data)
    inertia.append(kmeans.inertia_)

#  armazenando os valores de k e a inércia
elbow_df = pd.DataFrame({'K': range(1, 11), 'Inertia': inertia})

fig = px.line(elbow_df, x='K', y='Inertia', title='Método do Cotovelo para Determinar o Número Ideal de Clusters')
fig.update_traces(mode='lines+markers')
fig.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
data['cluster'] = kmeans.fit_predict(data)
```

```python
print(data['cluster'].value_counts())
# média de cada variável por cluster
cluster_means = data.groupby('cluster').mean()
print(cluster_means)
```

```python
import plotly.express as px

# Definindo rótulos personalizados para os clusters
cluster_labels = {0: 'Cluster 1', 1: 'Cluster 2', 2: 'Cluster 3'}

# gráfico de dispersão interativo com Plotly
fig = px.scatter(data, x='longitude', y='latitude', color='cluster', color_discrete_map=cluster_labels, title='Distribuição Espacial dos Clusters')
fig.update_traces(marker=dict(size=5), selector=dict(mode='markers'))
fig.update_layout(title_text='Distribuição Espacial dos Clusters', showlegend=True, autosize=False, width=800, height=700)
fig.show()
```
